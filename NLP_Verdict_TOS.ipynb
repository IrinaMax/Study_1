```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk import tokenize
#creating analyser will see how it will work
sid = SentimentIntensityAnalyzer()
```

```python
messages = df2[0]
for m in messages:
    vs = sid.polarity_scores(m)
    print("{:<70} {}".format(m, str(vs)))
    
```

```python
# creating and filling outcome table
data_all = df2[0]
messages = []
vs_compound = []
vs_pos = []
vs_neu = []
vs_neg = []

for i in range(0, len(data_all)):
    messages.append(data_all[i])
    vs_compound.append(sid.polarity_scores(data_all[i])['compound'])
    vs_pos.append(sid.polarity_scores(data_all[i])['pos'])
    vs_neu.append(sid.polarity_scores(data_all[i])['neu'])
    vs_neg.append(sid.polarity_scores(data_all[i])['neg'])
```

```python
dfsa = pd.DataFrame({'Messages': messages,
                        'Compound': vs_compound,
                        'Positive': vs_pos,
                        'Neutral': vs_neu,
                        'Negative': vs_neg})
dfsa = dfsa[['Messages', 'Compound', 'Positive', 'Neutral', 'Negative']]
```
