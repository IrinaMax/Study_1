```python
X = df[['like_per_month', 'newfriend_per_month',  'reply_per_month']]
y = df['is_churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=10).fit(X_train, y_train)
```

```python
rfc
# predict on test set
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
accuracy_score(y_test, rfc_pred)
```

```python
# f1 score must be  2 * (precision * recall) / (precision + recall)
f1_score(y_test, rfc_pred)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, rfc_pred))
```

```python
# recall score
from sklearn.metrics import recall_score

recall_score(y_test, rfc_pred)
```

```python
#  I will try oversampling minority class
# separate minority and majority classes
from sklearn.utils import resample
df['is_churn'] *=1   # move to numeric
df
```

```python
nch = X[X.is_churn==0]
ch = X[X.is_churn==1]
X.is_churn.value_counts()
```

```python
ch_upsampled = resample(ch,
                          replace=True, # sample with replacement
                          n_samples=len(nch), # match number in majority class
                          random_state=27) # reproducible results
ch_upsampled
```

```python
# combine majority and upsampled minority
upsampled = pd.concat([nch, ch_upsampled])
upsampled
```

```python
## check new class counts
upsampled['is_churn'].value_counts()
```
